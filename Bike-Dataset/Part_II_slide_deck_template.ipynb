## Investigation Overview


> The overall goal is to analyze data on the bikesharing system of the greater San Francisco Bay area. Key insights to be realised from the dataset include the distribution of Gender across bike trips, the system's popular user type, if there is any relationship between member age and duration on the road and lastly if bikers were comfortable sharing their bike trip details.




## Dataset Overview

> This dataset includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_FordGoBike = pd.read_csv('201902-fordgobike-tripdata.csv')
```

## The distribution of  Bike trips across Gender

> We would like to find out which gender was popular with the bike sharing system. From the plot, we can observe that the Male gender patronized the bikesharing system than Female and Other genders
 

```python
df_FordGoBike.member_gender.value_counts().plot.bar()
plt.title('Most Popular Gender')
plt.ylabel('Number of trips');
```

## The different user types popular with the bike sharing system

> The plot below depicts the user type that is popular with the system. It is observed that users preferred to be subscribers than customers.

```python
df_FordGoBike['user_type'].value_counts().plot(kind='bar');
plt.title('Most popular user type')
plt.ylabel('Number of trips');
```

## Relationship between member age and duration

There is a negative relationship between biker age and duration. The older a biker is, the less duration covered.

```python
# Create a new column for member ages.

df_FordGoBike['member_age'] = 2019- df_FordGoBike['member_birth_year']
```

```python
plt.scatter(data=df_FordGoBike, x='member_age', y='duration_sec')
plt.title('Member age vs Duration(sec)')
plt.xlabel('member_age')
plt.ylabel('duration(sec)');
```

## Are bikers comfortable sharing their  bike trip details?

Over 90% of the bikers were not comfortable to share their bike trip details.

```python
df_FordGoBike['bike_share_for_all_trip'].value_counts().plot(kind='bar');
plt.title('Bike Trip Share')
plt.ylabel('Number of trips');
```
